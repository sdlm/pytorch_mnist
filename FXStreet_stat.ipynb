```python
import pandas as pd

df = pd.read_json('./data.json')
```

## Feeds by company

```python
pd.value_counts(df['company']).plot.bar();
```

## Feeds by author

```python
pd.value_counts(df['author']).plot.bar();
```

## Feeds by tag
Show top 50 tags

```python
tags = set()
for i in range(14970):
    tags |= set(df.iloc[i]['tags'])

tag_with_feeds_count = {tag: 0 for tag in tags}
for i in range(14970):
    for tag in df.iloc[i]['tags']:
        tag_with_feeds_count[tag] += 1

tags_df = pd.DataFrame(tag_with_feeds_count.items(), columns=['tag', 'count'])

tags_df.sort_values(by=['count'], ascending=False, inplace=True)

tags_df[:50].plot(x='tag', y='count', kind='bar', figsize=(16, 3));
```

## Feeds by month
Sorted by feeds count in month

```python
df_timestamp = df.sort_values(by=['timestamp'])

df['month'] = df['timestamp'].apply(lambda x: str(x)[:7])

pd.value_counts(df['month']).plot.bar();
```
